# Introduction to Convolutions

This notebook runs through a few examples of how convolutions can be used to extract / augment information contained in an image.

It's meant to be an intuitive intro for how Convolutional Neural Networks (or ConvNets) use convolutions (or filters) to detect what's in an image.

```python
%matplotlib inline
from scipy.signal import convolve2d
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

im = plt.imread('../images/grid.jpg') # grayscale image
plt.imshow(im,cmap='gray')
```

```python
from scipy.signal import convolve2d

# vertical edge detector
# [1 0 -1]
# [1 0 -1]
# [1 0 -1]
vert_kernel = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
print(vert_kernel)
vertical = convolve2d(im, vert_kernel, 'same')

plt.imshow(vertical,cmap='gray')
```

```python
# horizontal edge detector
# [-1 -1 -1]
# [0  0  0]
# [1  1  1]
horiz_kernel = np.rot90(vert_kernel)
print(horiz_kernel)
horizontal = convolve2d(im, horiz_kernel, 'same')

plt.imshow(horizontal,cmap='gray')
```

```python
# Example edge detector
# [-1 -1 -1]
# [-1 8  -1]
# [-1 -1 -1]
edge_kernel = np.array([[-1, -1, -1],
                        [-1, 8, -1],
                        [-1, -1, -1]])
print(edge_kernel)
edge = convolve2d(im, edge_kernel, 'same')
plt.figure()
plt.imshow(edge,cmap='gray')
```

```python
# Example blur kernel
# [.0625 .125 .0625]
# [.125 .25  .125]
# [.0625 .125 .0625]
blur_kernel = np.array([[0.0625, 0.125, 0.0625],
                        [0.125, 0.25, 0.125],
                        [0.0625, 0.125, 0.0625]])
blur = convolve2d(im, blur_kernel, 'same')
plt.figure()
plt.imshow(blur,cmap='gray')
```
